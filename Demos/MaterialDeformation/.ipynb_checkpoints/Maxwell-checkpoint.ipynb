# Maxwell material deformation

This is a demonstration of the Maxwell model of viscoelastic material for EAS 4803/8803, Glacier and Ice Sheet Dynamics. Last updated by Alex Robel on Feb. 1, 2021. This is used in PS2

## Load Packages

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Parameters

```python
E = 1e10      #elastic modulus (in Pa) of ice
mu = 1e14     #linearized viscosity of ice 

ts = 3600*np.arange(24*4+1)             #time steps for simulations (hourly for 4 days)
period = 3600*24                       #period of forcing
sigma = 1e4*np.sin(2*np.pi*ts/period)  #a few days of diurnal tidal forcing
nt = np.size(ts)
dt = ts[1]-ts[0]                 #time step
```

## Viscoelastic Material Deformation Under Periodic Forcing

```python
#reminder of parameters
E = 1e10      #elastic modulus (in Pa) of ice
mu = 2e14     #linearized viscosity of ice 

#run viscoelastic model
strain_ve = np.zeros([nt+1])      #pre-allocate
strain_ve[0] = 0                  #initial condition
dsigma_dt = np.diff(sigma)/np.diff(ts)   #take time derivative of applied stress (sigma)

for t in range(nt-1):
    strainrate_ve = (sigma[t]/mu) + (dsigma_dt[t]/E)   #maxwell viscoelastic model
    strain_ve[t+1] = strain_ve[t] + strainrate_ve*dt   
```

```python
#plot
fig, ax1 = plt.subplots()

color = 'tab:red'
ax1.set_xlabel('time (days)')
ax1.set_ylabel('applied stress (Pa)', color=color)
ax1.plot(ts/(3600*24),sigma, color=color)
ax1.tick_params(axis='y', labelcolor=color)

ax2 = ax1.twinx()  # instantiate a second axes that shares the same x-axis

color = 'tab:blue'
ax2.set_ylabel('strain (unitless)', color=color)  # we already handled the x-label with ax1
ax2.plot(ts/(3600*24),strain_ve[0:-1] , color=color, linestyle=':')
ax2.tick_params(axis='y', labelcolor=color)
#plt.ylim(-1,1)

fig.tight_layout()  # otherwise the right y-label is slightly clipped
#plt.xlim(-500,0)
plt.show()
```
